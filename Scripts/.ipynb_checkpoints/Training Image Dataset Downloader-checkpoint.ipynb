```python
import urllib.request
import pandas as pd
```

```python
def download_image(url, file_path, file_name):
    full_path = file_path + file_name + '.jpg'
    urllib.request.urlretrieve(url, full_path)
    print("Fetching image ", file_name)
```

```python
tempdf = pd.read_csv('/Users/vishalkumar/Documents/Scripts/notSizeChartUrls.csv')
```

```python
urldf = tempdf.set_axis(['URL'], axis=1, inplace=False)
```

```python

for i in range(len(urldf)):
    try:
        url = urldf.loc[i, "URL"]
        file_name = ('Image'+ str(i))
        download_image(url, '/Users/vishalkumar/Documents/Scripts/Images/Not', file_name)
    except Exception:
        print("404 Not Found Exception")
        continue
```
