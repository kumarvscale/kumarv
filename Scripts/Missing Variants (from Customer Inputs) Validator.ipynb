```python
import pandas as pd
import numpy as np
```

```python
cust_df = pd.read_csv('test_catalogs.csv')
```

```python
cust_df["ReviewsPresent"]=""
print(cust_df)
```

```python
#iterate through each row of cust_df
for index, row in cust_df.iterrows():
    #download csv file from url present in PRODUCT_FEED_URL column
    df = pd.read_csv(row['product_feed_url'])
    #save product_feed_url in a variable
    pfu = row['product_feed_url']
    #check if reviews column is present in downloaded csv file
    if 'reviews' in df.columns:
        cust_df.loc[index,'ReviewsPresent'] = 'Yes'
        print(pfu,'Yes')
    else:
        cust_df.loc[index,'ReviewsPresent'] = 'No'
        print(pfu,'No')

#save cust_df to csv file
cust_df.to_csv('ReviewsCatalogs.csv', index=False)
```
