```python
import requests
import json
import pandas as pd
import re

def downloadjson(url):
    r = requests.get(url + '/products.json?limit=250')
    data = r.json()['products']
    return data

def parsejson(jsondata,url):
    products = []

    for product in jsondata:
        mainid = product['id']
        maintitle = product['title']
        product_type = product['product_type']
        brand = product['vendor']
        body_html = product['body_html']

        text = re.sub('<[^<]+?>', '', body_html)
        # Replace HTML character codes 
        text = text.replace("&nbsp;", " ")
        text = text.replace("&lt;", "<")
        text = text.replace("&gt;", ">")
        text = text.replace("&amp;", "&")
        text = text.replace("\n", "<br>")
        text = re.sub(r"(<br>)+", "<br>", text)
        description = text
        
        images = []
        for image in product['images']:
            try:
                image_link = image['src']
                element = f'<img src="{image_link}" width="300" height="300">'
                images.append(element)  # add the image to the list
                images_elements = '<br>'.join(images)
                # print(image)
            except:
                image = 'None'
                    
        for variant in product['variants']:
            


            item = {
                # 'Main_id': mainid,
                'title_data' : maintitle,
                # 'product_type' : product_type,
                'variant_id' : variant['id'],
                'variant_data' : variant['title'],
                'description_data' : description,
                # 'images_links' : images,
                'media_data':images_elements,
                'brand_data' : brand,
                'pdp_link' : url + 'products/' + product['handle'] + '?variant=' + str(variant['id'])
            }
            products.append(item)
    df = pd.DataFrame(products)
    return df

url = 'https://flamingo-demo-site.myshopify.com/'

raw_data = downloadjson(url)
df = parsejson(raw_data,url)
name = url.replace('https://', '')
name = name.replace('.com/','')
print(name)
df.to_csv(f'Shopify {name} data.csv',index=False)
df
```
