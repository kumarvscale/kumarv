```python
def is_valid_manufacturer_part_number(part_number):
    """
    Check if a given value is a valid manufacturer part number.
    """
    # Use regular expressions to validate the format of the part number
    pattern = "^[\\w\\-]{1,30}$"
    if not re.match(pattern, part_number):
        return False
    return True

#iterate through each row of adf
for index, row in adf.iterrows():
    #check if attribute_name is Manufacturer Part Number
    if row['attribute_name'] == 'manufacturerPartNumber':
        #check if attribute_value is valid
        if is_valid_manufacturer_part_number(row['attribute_value']):
            print('Valid Manufacturer Part Number')
            print(row['attribute_value'])
        else:
            print('Invalid Manufacturer Part Number')
            print(row['attribute_value'])

```
